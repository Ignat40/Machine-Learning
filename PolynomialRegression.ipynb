```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
ds = pd.read_csv('/content/sample_data/Position_Salaries.csv')
X = ds.iloc[:, 1:-1].values
y = ds.iloc[:, -1].values
```

Training the model


```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# converting the linear to polynomial model
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

Visualising the Linear Regression results

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title("Truth of Bluff")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
#visualising the polynomial regression model
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(X_poly), color = 'blue')
plt.title("Truth of Bluff (Polynomial)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

Linear Predicdtion

```python
lin_reg.predict([[6.5]])
```

Polynomial Prediciton

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
